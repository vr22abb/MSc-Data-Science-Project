```python
# Load the Datasets

import pandas as pd

daily_summaries = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MSc Data Science Project/ Daily Summaries - GHCND.csv')
monthly_summary = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MSc Data Science Project/Global Summary of the Month - GSOM.csv')
yearly_summary = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MSc Data Science Project/Global Summary of The Year - GSOY.csv')

#print("Daily Summaries")
#print(daily_summaries.head())

#print("\nGlobal Summary of the Month")
#print(monthly_summary.head())

#print("\nGlobal Summary of the Year")
#print(yearly_summary.head())

# Inspect the Data

print("Daily Summaries Info")
print(daily_summaries.info())

print("\nGlobal Summary of the Month Info")
print(monthly_summary.info())

print("\nGlobal Summary of the Year Info")
print(yearly_summary.info())

# Identify missing values:

print("Missing Values in Daily Summaries")
print(daily_summaries.isnull().sum())

print("\nMissing Values in Global Summary of the Month")
print(monthly_summary.isnull().sum())

print("\nMissing Values in Global Summary of the Year")
print(yearly_summary.isnull().sum())
```

```python
# Handle Missing Values

# Identify numeric and non-numeric columns
numeric_columns_daily = daily_summaries.select_dtypes(include=['number']).columns
non_numeric_columns_daily = daily_summaries.select_dtypes(exclude=['number']).columns

numeric_columns_monthly = monthly_summary.select_dtypes(include=['number']).columns
non_numeric_columns_monthly = monthly_summary.select_dtypes(exclude=['number']).columns

numeric_columns_yearly = yearly_summary.select_dtypes(include=['number']).columns
non_numeric_columns_yearly = yearly_summary.select_dtypes(exclude=['number']).columns

# Fill missing values in numeric columns with mean

daily_summaries[numeric_columns_daily] = daily_summaries[numeric_columns_daily].fillna(daily_summaries[numeric_columns_daily].mean())
monthly_summary[numeric_columns_monthly] = monthly_summary[numeric_columns_monthly].fillna(monthly_summary[numeric_columns_monthly].mean())
yearly_summary[numeric_columns_yearly] = yearly_summary[numeric_columns_yearly].fillna(yearly_summary[numeric_columns_yearly].mean())

# Fill missing values in non-numeric columns with mode
for column in non_numeric_columns_daily:
    daily_summaries[column] = daily_summaries[column].fillna(daily_summaries[column].mode()[0])

for column in non_numeric_columns_monthly:
    monthly_summary[column] = monthly_summary[column].fillna(monthly_summary[column].mode()[0])

for column in non_numeric_columns_yearly:
    yearly_summary[column] = yearly_summary[column].fillna(yearly_summary[column].mode()[0])

# Check for remaining missing values

print("Missing Values after filling in Daily Summaries")
print(daily_summaries.isnull().sum())

print("\nMissing Values after filling in Global Summary of the Month")
print(monthly_summary.isnull().sum())

print("\nMissing Values after filling in Global Summary of the Year")
print(yearly_summary.isnull().sum())
```
